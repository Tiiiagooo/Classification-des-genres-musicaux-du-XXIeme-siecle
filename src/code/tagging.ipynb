Ici l'objectif est de pouvoir attribuer un tag aux paroles de chansons.

```python
from function.utils import os, tqdm, json, glob

PATH_JSON = "../data/json_file/"
dictionnary1 = PATH_JSON+"dictionnaire_song.json"
dictionnary2 = PATH_JSON+"dictionnaire_song3.json"
```

```python
with open(dictionnary1, 'r') as f:
    dict1 = json.load(f)

with open(dictionnary2, 'r') as f:
    dict2 = json.load(f)    


len(dict2)
```

```python

for nom_artiste, songs in dict1.items():
    if nom_artiste in dict2.keys():
        for titre, lyrics, in songs.items():
            if titre not in dict2[nom_artiste]:
                dict2[nom_artiste][titre] = lyrics
            else:
                lyrics_dict1 = lyrics.get("lyrics")
                lyrics_dict2 = dict2[nom_artiste][titre].get("lyrics_primaire")

                if not lyrics_dict2 and lyrics_dict1:
                    dict2[nom_artiste][titre]["lyrics"] = lyrics_dict1

    else:
        dict2[nom_artiste] = songs

len(dict2)
```
